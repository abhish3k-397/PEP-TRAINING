Functions - Continue

```python
for i in range(5):
  print("*")
```

```python
for i in range(5):
  print("*****")
```

```python
for i in range(6):
  print("*"*i)
```

```python
n = int(input())
for i in range(n):
  print("*"*(n-i))
```

```python
n = int(input())
for i in range(n):
  for j in range(1,i+1):
    # print(" ", end="")
    print(j, end="")
  print()
```

```python
n = int(input())
for i in range(n,1,-1):
  for j in range(1,i):
    # print(" ", end="")
    print(j, end="")
  print()
```

```python
n = int(input())
for i in range(n):
  for j in range(n):
    print(i,end="")
  print()
```

```python
n = int(input())
for i in range(n):
  for j in range(i):
    print(chr(64+j+1),end="")
  print()
```

```python
n = int(input())
for i in range(n):
  for j in range(i):
    print(i,end="")
  print()
```

```python
n = int(input())
for i in range(n,0,-1):
  for j in range(n,n-i,-1):
    print(j,end="")
  print()
```

```python
n = int(input())
for i in range(1,n):
  for j in range(n-i,0,-1):
    print(j,end="")
  print()
```

```python
n = int(input())
for i in range(1,n):
  print("Next Building:")
  for j in range(1,n):
    for k in range(1,n):
      print(f"building: {i} , floor: {j}, room: {k} ")
    print("Next Floor")
  print()
```
