# File Handling in Python

**File handling** in Python is used to create, read, write, append, and delete files.

## 1. Opening a File

Use the `open()` function.

**Syntax:**

```python
file = open("filename.txt", "mode")
```

### File Modes

| Mode   | Description                                       |
| ------ | ------------------------------------------------- |
| `"r"`  | Read (default)                                    |
| `"w"`  | Write (creates a new file or overwrites existing) |
| `"a"`  | Append (adds data to the end of the file)         |
| `"x"`  | Create a new file (fails if file exists)          |
| `"rb"` | Read in binary mode                               |
| `"wb"` | Write in binary mode                              |

---

## 2. Reading a File

### Read the entire file

```python
file = open("sample.txt", "r")
print(file.read())
file.close()
```

### Read one line

```python
file = open("sample.txt", "r")
print(file.readline())
file.close()
```

### Read all lines

```python
file = open("sample.txt", "r")
print(file.readlines())
file.close()
```

---

## 3. Writing to a File

```python
file = open("sample.txt", "w")
file.write("Hello, Python!")
file.close()
```

**Output in `sample.txt`:**

```
Hello, Python!
```

> **Note:** `"w"` mode erases the existing content before writing.

---

## 4. Appending to a File

```python
file = open("sample.txt", "a")
file.write("\nWelcome to File Handling.")
file.close()
```

**Output:**

```
Hello, Python!
Welcome to File Handling.
```

---

## 5. Using `with` Statement (Recommended)

The `with` statement automatically closes the file.

```python
with open("sample.txt", "r") as file:
    data = file.read()
    print(data)
```

---

## 6. Checking if a File Exists

```python
import os

if os.path.exists("sample.txt"):
    print("File exists.")
else:
    print("File not found.")
```

---

## 7. Deleting a File

```python
import os

if os.path.exists("sample.txt"):
    os.remove("sample.txt")
    print("File deleted.")
else:
    print("File does not exist.")
```

---

## 8. Exception Handling

```python
try:
    with open("sample.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found.")
```

## Summary

| Operation       | Method                                |
| --------------- | ------------------------------------- |
| Open file       | `open()`                              |
| Read file       | `read()`, `readline()`, `readlines()` |
| Write file      | `write()`                             |
| Append data     | `write()` with `"a"` mode             |
| Close file      | `close()` or use `with`               |
| Delete file     | `os.remove()`                         |
| Check existence | `os.path.exists()`                    |

**Best Practice:** Use the `with` statement for file handling because it automatically closes the file and helps prevent resource leaks.
