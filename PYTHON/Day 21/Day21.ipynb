```python
s_per = pd.read_csv("/content/LPU-Training-Resources/Python/Day_22_23_Pandas/student_performance_project.csv")
```

```python
s_per
```

```python
s_per.isnull().sum()
```

```python
# s_per.dropna(ignore_index=True)
for col in ['Math', 'Science', 'English']:
    if s_per[col].isnull().any():
        s_per[col] = s_per[col].fillna(s_per[col].mean().round(2))
s_per.isnull().sum()
```

```python
s_per.duplicated().sum()
```

```python
s_per.drop_duplicates(inplace=True)
# s_per.duplicated().sum()
```

```python
s_per.duplicated()
```

```python
stud_from_del = s_per[s_per['City'] == 'Delhi']
display(stud_from_del.head())
```

```python
display(s_per[['Math','English','Science']].agg(['mean','max']))
```

```python
total = s_per['Math'] + s_per['Science'] + s_per['English']
s_per['Total'] = total
display(s_per.head())
```

```python
average = s_per['Total']/3
s_per['Average'] = average
display(s_per.head())
```

```python
rank = s_per['Average'].rank(ascending=False,method='min')
s_per['Rank'] = rank
display(s_per.sort_values('Rank'))
```

```python
s_per['Result'] = s_per[['Math','English','Science']].apply(lambda x: 'Pass' if x.mean() > 50 else 'Fail',axis=1)
display(s_per)
```

```python
gen = s_per.groupby(['Gender'])[['Math','English','Science','Average']].mean()
display(gen)
```

```python
Topper = s_per.sort_values('Rank').head(10)
display(Topper)
```

```python
s_per.to_csv('student_performance_analysis.csv', index=False)
```
