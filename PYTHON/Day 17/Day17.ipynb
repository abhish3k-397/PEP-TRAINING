```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
  @abstractmethod
  def fuel_type(self):
    pass


class PetrolCar(Vehicle):
  def fuel_type(self):
    print("Petrol Car")


class DieselCar(Vehicle):
  def fuel_type(self):
    print("Diesel Car")


class ElectricCar(Vehicle):
  def fuel_type(self):
    print("Electric Car")

Car1 = PetrolCar()
Car2 = DieselCar()
Car3 = ElectricCar()
Car4 = PetrolCar()
Car5 = DieselCar()

Car1.fuel_type()
Car2.fuel_type()
Car3.fuel_type()
Car4.fuel_type()
Car5.fuel_type()
```

```python
class Person:
  def __init__(self, name, age):
    super().__init__()
    self.name = name
    self.age = age

  def greet(self):
    print("Hello my name is " + self.name)
class Teacher(Person):
  def __init__(self, name, age, subject):
    super().__init__(name, age)
    self.subject = subject

  def greet(self):
    print("Hello my name is " + self.name + " and I teach " + self.subject)
class Student(Person):
  def __init__(self,name,age,student_id):
    super().__init__(name,age)
    self.student_id = student_id

  def greet(self):
    print("Hello my name is " + self.name + " and my student id is " , self.student_id)


# class StudentTeacher(Student,Teacher):
#   def __init__(self,name,age,student_id,subject):
#     Student.__init__(self,name,age,student_id)
#     Teacher.__init__(self,name,age,subject)

#   def greet(self):
#     print("Hello my name is " + self.name + " and I teach " + self.subject + " and my student id is " , self.student_id)
T1 = Teacher("Abhi",21,"English")
T2 = Teacher("Arjun",43,"Python")
S1 = Student("Arun",16,123212)
S2 = Student("AK",17,123232)
T1.greet()
T2.greet()
S1.greet()
S2.greet()
# ST1 = StudentTeacher("Arun",16,123212,"Python")
# ST1.greet()
```

```python
class Book:
  def __init(self,book_id,title,author):
    self.book_id = book_id
    self.title = title
    self.author = author
    self.__available = True
  def display(self):
    print(self.book_id,self.title,self.author)
  def borrow(self):
    if self.__available:
      self.__available = False
      print("Book borrowed")
    else:
      print("Book not available")
  def return(self):
    if not self.__available:
      self.__available = True
      print("Book returned")
    else:
      print("Book not borrowed")
```
