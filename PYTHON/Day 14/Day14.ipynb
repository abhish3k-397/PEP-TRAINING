```python
social_network = {
    "Abhi": {"Rahul", "Aman", "Priya"},
    "Rahul": {"Abhi", "Priya", "Riya"},
    "Aman": {"Abhi", "Riya"},
    "Priya": {"Abhi", "Rahul"},
    "Riya": {"Rahul", "Aman"}
}

user = "Abhi"
friend = "Rahul"

# Mutual friends
mutual = social_network[user] & social_network[friend]

# Friend suggestions
suggestions = social_network[friend] - social_network[user] - {user}

print("Mutual Friends:", mutual)
print("Friend Suggestions:", suggestions)
```

```python
purchase_history = {
    "Abhi": {"PC", "Laptop", "Mouse"},
    "Rahul": {"Mouse", "Keyboard", "Camera"},
    "Aman": {"Phone", "Camera"},
    "Priya": {"PC", "Camera"},
    "Riya": {"Headphones", "Mouse"}
}

user = input("username: ")
others = set()

for name, products in purchase_history.items():
    if name != user:
        others |= products

not_purchased = others - purchase_history[user]

print("Products purchased by others but not by", user, ":", not_purchased)
```

```python
def display():
  print("Name\t\tRank")
  for name, rank in ranks.items():
    print(f"{name}\t\t{rank}")

def refresh(ranks):
  return dict(sorted(ranks.items() , key = lambda x:x[1]))

ranks = {}
name = input("Enter the name or \"end\" to stop: ")
while name != "end":
  rank = int(input("Enter the rank: "))
  ranks[name] = rank
  ranks = refresh(ranks)
  display()
  name = input("Enter the name or \"end\" to stop: ")
```

```python
assignment1 = input("Enter Assignment 1: ").lower().split()
assignment2 = input("Enter Assignment 2: ").lower().split()

set1 = set(assignment1)
set2 = set(assignment2)

common_words = set1 & set2
all_unique_words = set1 | set2

similarity = (len(common_words) / len(all_unique_words)) * 100

print("\nUnique words in Assignment 1:", set1)
print("Unique words in Assignment 2:", set2)
print("Common words:", common_words)
print(f"Similarity Percentage: {similarity:.2f}%")

if similarity >= 80:
    print("High chance of plagiarism")
elif similarity >= 50:
    print("Moderate similarity")
else:
    print("Low similarity")
```
