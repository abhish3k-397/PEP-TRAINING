Q1 Area Calculator

```python
def area(l,b):
  return l*b
l = int(input("Enter length : "))
b = int(input("Enter breadth: "))
print(area(l,b))
```

Q2 SI calculator

```python
def si(p,r,t):
  return (p*r*t)/100
p = int(input("Enter principle amount : "))
r = int(input("Enter rate of interest : "))
t = int(input("Enter time : "))
print(si(p,r,t))
```

Q3 Celcius to Fahrenheit

Q6 password verification

```python
def verifypass(username,password,stored_username,stored_password):
  if username == stored_username and password == stored_password:
    return "Login Successful"
  else:
    return "Login Failed Username or password incorrect"
stored_username = "abhi"
stored_password = "krish"
username = input("Enter username : ")
password = input("Enter password : ")
print(verifypass(username,password,stored_username,stored_password))

```

Q7 Create family registery

```python
def info(data):
  for key in data:
    print(f"{key} : {data[key]}")
data = {}
while True:
  relation = input("Enter Relation or \"end\" to stop: ")
  if relation == "end":
    break
  name = input("Enter Name: ")
  data[relation] = name
info(data)
```

Q8 bill generator

```python
def bill(data):
  total = 0
  for key in data:
    print(f"{key} : {data[key]}")
    total += data[key]
  print(f"Total : {total}")

data = {}
while True:
  name = input("Enter name of the product or \"end\": ")
  if name == "end":
    break
  price = int(input("Enter price of the product: "))
  discount = int(input("Enter discount if applicable in percentage otherwise 0: "))
  discounted_price = price * ((100 - discount)/100)
  data[name] = discounted_price
bill(data)
```

Q9 Total Marks

```python
def marksheet(marks):
  total = 0
  for key in marks:
    print(f"{key} : {marks[key]}")
    total += marks[key]
  print(f"Total Marks: {total}")
  percentage = (total/len(marks))
  print(f"Percentage: {percentage}%")
marks = {}
while True:
  name = input("Enter name of the Subject or \"end\" to stop: ")
  if name == "end":
    break
  marks[name] = int(input("Enter marks: "))
marksheet(marks)
```

Q10 salary calculator

```python
def salary(salary):
  for key in salary:
    print(f"{key} : {salary[key]}")
salary = {}
while True:
  name = input("Enter name of the Employee or \"end\" to stop: ")
  if name == "end":
    break
  sal = int(input("Enter Salary: "))
  HRA = int(input("Enter HRA: "))
  DA = int(input("Enter DA: "))
  salary[name] = sal + HRA + DA
salary(salary)
```
