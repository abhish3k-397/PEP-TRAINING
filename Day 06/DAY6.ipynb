```python
def greet(name, message = "Good Morning"):
  print(f"Hi, {name}. {message}")
greet("Abhi","Hello")
greet("Abhi")
name = input()
greet(name)
```

```python
def calc(a,b,op):
  match op:
    case "+":
      return a+b
    case "-":
      return a-b
    case "*":
      return a*b
    case "/":
      return a/b
    case "**":
      return a**b
    case "***":
      return pow(a,3)
    case "//":
      return a//b
    case "%":
      return a%b
    case _:
      print("Invalid Operation")
      return
a = int(input("Enter value a: "))
op = input("Enter the operatio")
b = int(input("Enter value b: "))
result = calc(a,b,op)
print(f"{a} {op} {b} = {result}")
```

Largest number


```python
def largest(a,b):
  return a if a>b else b
a = int(input("Enter value a: "))
b = int(input("Enter value b: "))


largest(a,b)
```

```python
def largest5(a,b,c,d,e):
  return max(a,b,c,d,e)
a = int(input("Enter value a: "))
b = int(input("Enter value b: "))
c = int(input("Enter value c: "))
d = int(input("Enter value d: "))
e = int(input("Enter value e: "))
largest5(a,b,c,d,e)
```

```python
def large(*args):
  return max(args)

a = int(input("Enter value a: "))
b = int(input("Enter value b: "))
c = int(input("Enter value c: "))
d = int(input("Enter value d: "))
e = int(input("Enter value e: "))
large(a,b,c,d,e)
```
