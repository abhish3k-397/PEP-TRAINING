```python
names = []
name = input("Enter customer name: ")
while name != "end":
    names.append(name)
    name = input("Enter customer name: ")
for i in names:
  print(i.title())
```

```python
email = input("Enter email: ")
domain = email.split("@")[1]
print(domain)
```

```python
countAlpha = 0
countSpace = 0
countDigit = 0
countSpeci = 0
string = input("Enter string: ")
for i in string:
  if i.isalpha():
    countAlpha += 1
  elif i.isdigit():
    countDigit += 1
  elif i.isspace():
    countSpace += 1
  else:
    countSpeci += 1
print("Alpha: ",countAlpha)
print("Space: ",countSpace)
print("Digit: ",countDigit)
print("Speci: ",countSpeci)
```

```python
from collections import Counter
str1 = input("Enter String 1: ")
str2 = input("Enter String 2: ")
c1 = Counter(str1)
c2 = Counter(str2)
print(c1)
print(c2)
if c1 == c2:
  print("Anagram")
else:
  print("Not Anagram")
```

```python
sent = input("Enter your sentence")
words = sent.split()
longest = ""
for i in words:
  if len(i) > len(longest):
    longest = i
print("longest word: ",longest)
```

```python
sent = input("Enter your sentence")
sent = sent.replace(" ","_")
print(sent)
```
